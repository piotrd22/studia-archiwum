```python
import pandas
import matplotlib.pyplot as plt
```

```python
miasta = pandas.read_csv("./miasta.csv")
miasta.values
```

```python
miasta.loc[len(miasta)] = [2010,460,555,405]
miasta
```

```python
years = miasta["Rok"]
gd = miasta["Gdansk"]
plt.plot(years, gd, color="r", marker="o", linewidth=2.0)
plt.title("Ludnosc w mieście Gdańsk")
plt.xlabel("Lata")
plt.ylabel("Liczba ludnosci (w tys.)")
plt.show()
```

```python
years = miasta["Rok"]
gd = miasta["Gdansk"]
poz = miasta["Poznan"]
sz = miasta["Szczecin"]

plt.plot(years, gd, color="r", marker="o", linewidth=2.0, label="Gdańsk")
plt.plot(years, sz, color="b", marker="o", linewidth=2.0, label="Szczecin")
plt.plot(years, poz, color="y", marker="o", linewidth=2.0, label="Poznań")

plt.title("Ludnosc w miastach Polski")
plt.xlabel("Lata")
plt.ylabel("Liczba ludnosci (w tys.)")
plt.legend()
plt.show()
```

```python
def standarize(data):
    std = data.std()
    mean = data.mean()
    return data.map(lambda x: (x - mean) / std)

print("Cities", "Mean", "Std")
for i in ["Gdansk", "Szczecin", "Poznan"]:
   standarised = standarize(miasta[i])
   print(i, standarised.mean(), standarised.std()) 
```

```python
def normalize(data):
    minimum = data.min()
    maximum = data.max()
    return data.map(lambda x: (x - minimum) / (maximum - minimum))

print("Cities", "Min", "Max")
for i in ["Gdansk", "Szczecin", "Poznan"]:
   normalized = normalize(miasta[i])
   print(i, normalized.min(), normalized.max()) 
```
